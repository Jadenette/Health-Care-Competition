### Stats & Data Visualization

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('training_cleaning_jade.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
# drop missing genders and according rows
data = data.dropna(how='any', subset=['gender'])
data.shape
```

## drop all version

```python
corr = data.corr()
```

```python
corr.style.background_gradient()
```

```python
data_drop = data.dropna()
data_drop.shape
```

```python
# change the categorical data to label code

# ethnicity
data_drop['ethnicity'].astype(str)
data_drop['ethnicity'].unique() 
```

```python
# label_encoder object knows how to understand word labels. 
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data_drop['ethnicity'] = label_encoder.fit_transform(data_drop['ethnicity']) 
data_drop['ethnicity'].unique()
```

```python
# gender
data_drop['gender'].astype(str)
data_drop['gender'].unique() 
```

```python
data_drop['gender'] = label_encoder.fit_transform(data_drop['gender']) 
data_drop['gender'].unique()
```

```python
# hospital_admit_source        
data_drop['hospital_admit_source'].astype(str)
data_drop['hospital_admit_source'].unique() 
```

```python
data_drop['hospital_admit_source'] = label_encoder.fit_transform(data_drop['hospital_admit_source']) 
data_drop['hospital_admit_source'].unique()
```

```python
# icu_admit_source 
data_drop['icu_admit_source'].astype(str)
data_drop['icu_admit_source'].unique() 
```

```python
data_drop['icu_admit_source'] = label_encoder.fit_transform(data_drop['hospital_admit_source']) 
data_drop['icu_admit_source'].unique()
```

```python
# icu_stay_type
data_drop['icu_stay_type'].astype(str)
data_drop['icu_stay_type'].unique()              
```

```python
data_drop['icu_stay_type'] = label_encoder.fit_transform(data_drop['icu_stay_type']) 
data_drop['icu_stay_type'].unique()
```

```python
# icu_type
data_drop['icu_type'].astype(str)
data_drop['icu_type'].unique()              
```

```python
data_drop['icu_type'] = label_encoder.fit_transform(data_drop['icu_stay_type']) 
data_drop['icu_type'].unique()
```

```python
corr_dropall = data_drop.corr()
corr_dropall.style.background_gradient()
```
